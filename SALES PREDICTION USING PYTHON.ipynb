The company has conducted advertising campaigns and the sales for each campaign is given

I am going to use Linear Regression for sales prediction and I am also going to give a suggestion of a particular advertising campaign for the company to focus on !

The plot for Sales is more Linear for TV campaign and the other campaigns (Radio and Newspaper) have many outliers

```python
X=ad[['TV','Radio','Newspaper']]
Y=ad['Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

# MODEL CREATION

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

# MODEL FITTING

```python
model.fit(x_train,y_train)
```

# MODEL PREDICTIONS

```python
model.score(x_test,y_test)
```

Accuracy is 92%

```python
Y=model.predict(x_test)
```

```python
plt.scatter(y_test,Y)
plt.title("ACTUAL vs PREDICTED VALUES")
plt.xlabel("ACTUAL VALUES")
plt.ylabel("PREDICTED VALUES")
```

                               Predicted values are ON PAR with Actual values !
